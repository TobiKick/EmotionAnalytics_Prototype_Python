```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
```

```python
faces = np.load('faces_sequenced_45.npy', allow_pickle=True)
labels = np.load('labels_sequenced_45.npy', allow_pickle=True)
```

```python
print(faces.shape)
print(labels.shape)
labels =  labels.astype(np.float)
print(labels.min())
print(labels.max())
```

```python
fold = 0
subject = 5
for i in range(0, 45):
    img = faces[fold][subject][i]
    plt.imshow(img)
    plt.figure(i+1)
    print(labels[fold][subject][i])
```
